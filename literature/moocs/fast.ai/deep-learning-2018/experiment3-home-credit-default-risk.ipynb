## 2. Prepare dataset

### Aggregate and join extra data

Below function performs aggregation and joining of additional datasets. A lot of the feature engineering ideas were taken from https://www.kaggle.com/shep312/lightgbm-with-weighted-averages-dropout-771/code

#### Previous applications

Count the number of previous applications and get the total credit amount for each.

```python
prev_app_agg = previous_application.groupby('SK_ID_CURR').agg(
    {'SK_ID_CURR': 'count', 'AMT_CREDIT': 'sum'})
```

```python
prev_app_agg.columns = ['PREV APP COUNT', 'TOTAL PREV LOAN AMT']
```

```python
prev_app_agg.head()
```

```python
train_df = application_train.merge(
    prev_app_agg, left_on='SK_ID_CURR', right_index=True, how='left')
```

```python
test_df = application_test.merge(
    prev_app_agg, left_on='SK_ID_CURR', right_index=True, how='left')
```

Add the mean of all previous application features to the set.

```python
prev_app_mean = previous_application.groupby('SK_ID_CURR').mean()

train_df = train_df.merge(
    prev_app_mean, left_on='SK_ID_CURR', right_index=True,
    how='left', suffixes=['', '_PAVG'])
test_df = test_df.merge(
    prev_app_mean, left_on='SK_ID_CURR', right_index=True,
    how='left', suffixes=['', '_PAVG'])
```

Get the mode of each categorical variable.

```python
categorical_feats = previous_application.columns[
    previous_application.dtypes == 'object']
```

```python
def get_mode(x):
    x = x.fillna('NaN')
    return x.mode().iloc[0]
```

```python
%time prev_apps_cat_mode = previous_application[list(categorical_feats) + ['SK_ID_CURR']].groupby('SK_ID_CURR').agg({k: get_mode for k in categorical_feats})
```

```python
train_df = train_df.merge(
    prev_apps_cat_mode, left_on='SK_ID_CURR',
    right_index=True, how='left', suffixes=['', '_BMODE'])
test_df = test_df.merge(
    prev_apps_cat_mode, left_on='SK_ID_CURR',
    right_index=True, how='left', suffixes=['', '_BMODE'])
```

#### Credit card balance

Add credit card balance weighted by recency.

```python
def get_weighted_avg(x):
    return np.average(x, weights=-1/credit_card_balance.loc[x.index, 'MONTHS_BALANCE'])
```

```python
credit_card_avgs = credit_card_balance.groupby('SK_ID_CURR').agg(get_weighted_avg)
```

```python
train_df = train_df.merge(
    credit_card_avgs, left_on='SK_ID_CURR', right_index=True,
    how='left', suffixes=['', '_CCAVG'])
test_df = test_df.merge(
    credit_card_avgs, left_on='SK_ID_CURR', right_index=True,
    how='left', suffixes=['', '_CCAVG'])
```

Add credit card categorical data (using most recent category).

```python
most_recent_index = credit_card_balance.groupby('SK_ID_CURR')['MONTHS_BALANCE'].idxmax()
cat_feats = credit_card_balance.columns[credit_card_balance.dtypes == 'object'].tolist()  + ['SK_ID_CURR']
```

```python
train_df = train_df.merge(
    credit_card_balance.loc[most_recent_index, cat_feats],
    left_on='SK_ID_CURR', right_on='SK_ID_CURR',
    how='left', suffixes=['', '_CCAVG'])
test_df = test_df.merge(
    credit_card_balance.loc[most_recent_index, cat_feats],
    left_on='SK_ID_CURR', right_on='SK_ID_CURR',
    how='left', suffixes=['', '_CCAVG'])
```

#### Credit bureau data

```python
credit_bureau_avgs = bureau.groupby('SK_ID_CURR').mean()
```

```python
train_df = train_df.merge(
    credit_bureau_avgs, left_on='SK_ID_CURR', right_index=True,
    how='left', suffixes=['', '_BAVG'])
test_df = test_df.merge(
    credit_bureau_avgs, left_on='SK_ID_CURR', right_index=True,
    how='left', suffixes=['', '_BAVG'])
```

#### POS cash sales

```python
def get_weighted_avg(x):
    return np.average(x, weights=-1/pos.loc[x.index, 'MONTHS_BALANCE'])
```

```python
f = {
    'CNT_INSTALMENT': get_weighted_avg,
    'CNT_INSTALMENT_FUTURE': get_weighted_avg,
    'SK_DPD': get_weighted_avg,
    'SK_DPD_DEF': get_weighted_avg}

cash_avg = pos.groupby('SK_ID_CURR')[
    'CNT_INSTALMENT','CNT_INSTALMENT_FUTURE', 'SK_DPD', 'SK_DPD_DEF'].agg(f)
```

```python
train_df = train_df.merge(
    cash_avg, left_on='SK_ID_CURR', right_index=True,
    how='left', suffixes=['', '_CAVG'])
test_df = test_df.merge(
    cash_avg, left_on='SK_ID_CURR', right_index=True,
    how='left', suffixes=['', '_CAVG'])
```

#### POS cash balance - categorical features (most recent)

```python
most_recent_index = pos.groupby('SK_ID_CURR')['MONTHS_BALANCE'].idxmax()
cat_feats = pos.columns[pos.dtypes == 'object'].tolist()  + ['SK_ID_CURR']
```

```python
train_df = train_df.merge(
    pos.loc[most_recent_index, cat_feats],
    left_on='SK_ID_CURR', right_on='SK_ID_CURR',
    how='left', suffixes=['', '_CAVG'])
test_df = test_df.merge(
    pos.loc[most_recent_index, cat_feats],
    left_on='SK_ID_CURR', right_on='SK_ID_CURR',
    how='left', suffixes=['', '_CAVG'])
```

#### Installments payments (average)

```python
installments_avg = installments_payments.groupby('SK_ID_CURR').mean()
```

```python
train_df = train_df.merge(
    installments_avg, left_on='SK_ID_CURR', right_index=True,
    how='left', suffixes=['', '_IAVG'])
test_df = test_df.merge(
    installments_avg, left_on='SK_ID_CURR', right_index=True,
    how='left', suffixes=['', '_IAVG'])
```
